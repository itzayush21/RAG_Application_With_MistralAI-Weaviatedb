```python

from langchain.vectorstores import Weaviate
import weaviate
from weaviate.classes.init import Auth
client = weaviate.connect_to_wcs(
    cluster_url=WEAVIATE_CLUSTER,
    auth_credentials=Auth.api_key(WEAVIATE_API_KEY)
)
```

```python

# specify embedding model (using huggingface sentence transformer)
from langchain.embeddings import HuggingFaceEmbeddings
embedding_model_name = "sentence-transformers/all-mpnet-base-v2"
#model_kwargs = {"device": "cuda"}
embeddings = HuggingFaceEmbeddings(
  model_name=embedding_model_name,
  #model_kwargs=model_kwargs
)
```

```python
from langchain.document_loaders import PyPDFLoader
loader = PyPDFLoader("/content/rag.pdf", extract_images=True)
pages = loader.load()
```

## Splitting in Chucks


```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=20)
docs = text_splitter.split_documents(pages)
```

```python
!pip install weaviate-client>=4.0.0
```

```python
!pip install langchain-weaviate # Install the missing package 'langchain-weaviate' using pip.
```

```python
from langchain_weaviate.vectorstores import WeaviateVectorStore
vector_db = WeaviateVectorStore.from_documents(
    docs, embeddings, client=client, by_text=False
)
```

```python
print(vector_db.similarity_search("what is rag?", k=3)[0].page_content)
```

```python
from langchain.prompts import ChatPromptTemplate

template="""You are an assistant for question-answering tasks.
Use the following pieces of retrieved context to answer the question.
If you don't know the answer, just say that you don't know.
Use ten sentences maximum and keep the answer concise.
Question: {question}
Context: {context}
Answer:
"""
```

```python
prompt=ChatPromptTemplate.from_template(template)
```

```python

model = HuggingFaceHub(
    huggingfacehub_api_token="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",
    repo_id="mistralai/Mistral-7B-Instruct-v0.2",
    model_kwargs={"temperature":1, "max_length":180}
)
```

```python
from langchain.schema.runnable import RunnablePassthrough
from langchain.schema.output_parser import StrOutputParser
```

```python
output_parser=StrOutputParser()
retriever=vector_db.as_retriever()
```

```python
rag_chain = (
    {"context": retriever,  "question": RunnablePassthrough()}
    | prompt
    | model
    | output_parser
)
```

```python
print(rag_chain.invoke("what is rag system?"))
```

```python
print(rag_chain.invoke("How does the RAG model differ from traditional language generation models?"))
```
